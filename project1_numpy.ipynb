## Project 1: Student Marks Analysis Using NumPy

- Project Objective

The objective of this project is to analyze student marks data using the NumPy library.
I perform statistical analysis, sorting, condition-based analysis, and insights generation on a 1D NumPy array representing student marks.

#### Datase Description

The dataset contains marks of 12 students in a subject

```python
import numpy as np

stud_marks = np.array([80, 70, 45, 60, 30, 35, 90, 78, 50, 40, 35, 28])

print(stud_marks)
```

#### Basic information about dataset


Total Numbers of Students

```python
stud_marks.size
```


 Total number of students = 12

#### Data Type of Marks

```python
stud_marks.dtype
```

- All marks are stored as integer values
- Datasets is clean and consitent 

### Statistical Analysis of Student marks

Minimum Marks

```python
print(np.min(stud_marks))
```

- Minimum marks scored = 28
- Indicates lowest performing student

Maximum marks

```python
print(np.max(stud_marks))
```

- Maximum marks scored = 90
- Indicated highest performing student

Average (Mean) marks

```python
print(np.mean(stud_marks))
```

- Average marks ~ 53
- It shows overall class performance is average

Median marks

```python
print(np.median(stud_marks))
```

- Median Marks = 47.5
- Indicates that half of students scored below 47.5

Standard Deviation

```python
print(np.std(stud_marks))
```

- Standard deviation is high
- Marks are widely spread, showing performance gap among students

### Performance Based Analysis

Students Who Scored Less than 40 marks( Weak Performance)

```python
print(stud_marks[stud_marks<40])
```

- These students need extra academic support

- Indicates learning gaps

Students who Scored above 40 and below 75 marks (Average Performance)

```python
print(stud_marks[(stud_marks>40) & (stud_marks < 75)])
```

- These students performance is average 
- Teachers should focus on these students and students should focus more on their study

Students who scored above 75 (Good performance)

```python
print(stud_marks[stud_marks>=75])
```

- These students are performing well

- These are top 3 students of a class

### Pass & Fail Analysis

Pass Students

```python
pass_stud = stud_marks[stud_marks>=35]

print(pass_stud)

print(pass_stud.size)
```

- Students who scored 35 or above are considered passed
- Total Pass Students = 10
- 10 students are getting pass in the exam out of 12 students


Fail Students


```python
fail_stud = stud_marks[stud_marks<35]

print(fail_stud)

print(fail_stud.size)
```

- Students who scored below 35 are considered failed
- Total Fail Students = 2
- These students need special academic attention


### Class Passing Rate

```python
passing_rate = (pass_stud.size / stud_marks.size)*100

print(passing_rate)
```

- Class passing rate = 83.33 %

- Shows overall academic success of the class

### Class Failure Rate

```python
failure_rate = (fail_stud.size / stud_marks.size)* 100

print(failure_rate)
```

- Class failure rate = 16.66 %

### Key Observations From Pass and Fail analysis

- Majority of students have cleared the exam

- A small percentage of students have failed

- Passing rate reflects average to good class performance

- Failure rate suggests need for remedial classes

### Sorting Analysis

Marks Sorted in Ascending Order

```python
print( np.sort(stud_marks))
```

Top 3 Students Marks


```python
print(np.sort(stud_marks)[-3:])
```

- Identifies top-performing students
- useful for rewards and recognition
